```python

import pandas as pd
pd.options.display.float_format = '{:.4f}'.format
df1 = pd.read_csv("https://raw.githubusercontent.com/"
                  "panda-kim/pandas/main/shift01%EA%B0%95_df1.csv")
```

```python
df1["shift"] = df1["Close"].shift(1)
df1["등락률"] = (df1["Close"] - df1["shift"]) /df1["shift"]
df1
```

```python
df1["등락률2"] = df1["Close"].pct_change(1)
df1
```
