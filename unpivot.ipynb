```python
import pandas as pd
val = [[83, 93, 99], [77, 88, 91], [92, 78, 95]]
idx = ["권보아", "김범수", "박효신"]
df = pd.DataFrame(val, index=idx, columns=["국어", "수학", "과학"])
val1 = [[20, 10], [5, 25], [22, 8], [12, 18]]
idx1 = [["1반", "1반", "2반", "2반"], ["남", "여","남","여"]]
df1 = pd.DataFrame(val1,index=idx1, columns=["도시락", "샌드위치"])
df2 = df1.copy().unstack()
df3 = df1.copy()
df3["업체"] = ["CJ푸드", "신세계", "신세계", "CJ푸드"]
df4 = df3.reset_index().rename(columns={"level_0":"반", "level_1":"성별"}).copy()
```

```python

# 인덱스가 두줄, 컬럼이 없는 시리즈가 나옴
# 언피벗된 데이터프레임으로 만들려면 익덱스를 values로 만든다. 
# reset_index

df = df.stack().reset_index()
df.columns = ["이름","과목","점수"]
df
```

```python
#멀티인덱스

df1= df1.stack().reset_index()
df1.columns = ["반이름","성별","항목","수량"]
df1
```

```python
df2 = df2.stack(level=[0,1]).reset_index()
df2.columns = ["반","제품","성별","수량"]
df2
#level 0이 위 
```

```python
df3 = df3.set_index("업체",append = True).stack().reset_index()
df3.columns = ["반","성별", "업체","품목","수량"]
df3

```

## 언피벗 과정
1. set_index로 언피벗하지 않을 열을 인덱스로 만든다.
2. stack으로 칼럼을 인덱스로 보내 언피벗된 시리즈로 만들고 reset_index로 언피벗된 데이터프레임을 만든다. 
3. df.columns 혹은 rename으로 column들의 이름을 바꾼다. 

```python
df4.melt(id_vars=["반","성별","업체"], var_name = "제품", value_name = "수량")
```
