- 데이터셋은 분석에 적합하지 않은 형태로 주어지기도 한다.
- 단일 행을 추출하기는 쉽지만 데이터를 집계하기 어려운 형식으로 값을 저장하는 등 데이터를 벗어나 더 큰 구조적인 문제가 있는 경우도 있다.
- 데이터셋을 재구성한다는 것은 기존 형태에서 얻어내기 어려운 정보를 얻을 수 있도록 데이터셋을 다른 형태로 변환하는 것을 의미한다.
- 재구성은 데이터를 바라보는 새로운 관점을 제공하낟. 
- 데이터 분석의 80%는 데이터를 정리하고 적절한 형태로 변형하는 것이다.

- 데이터가 가로로 넓다는 것은 어떤 것을 의미할까?
- 넓은 데이터셋은 전체 그림을 볼 때 유용하다. 하지만 데이터셋은 열을 추가할수록 작업이 어려워진다.
- 좁은 데이터셋은 수직으로 확장된다. 좁은 형식을 사용하면 기존 데이터를 조작하고 새로운 레코드를 추가하기 쉽다. 각 변수는 별개의 열로 분리된다.

# 피벗테이블
- 피벗 테이블은 열의 값을 집계하고 다른 열의 값을 사용하여 결과를 그룹화한다.
- 집계라는 단어는 여러 값을 요약하는 계산을 나타낸다.
- 예를 들어 평균, 합계, 중앙값 및 개수가 집계에 포함된다.

1. 값을 집계할 열을 선택한다.
2. 열에 적용할 집계 연산을 선택한다.
3. 값이 집계된 데이터를 범주로 그룹화할 열을 선택한다.
4. 그룹을 행,축, 열 또는 두 축 모두에 배치할지 여부를 결정한다.

```python
sales
```

```python
# "숫자"열에 기본 집계 연산인 평균을 적용한다.
sales.pivot_table(index = "Date")
```

```python
sales.pivot_table(index = "Date", aggfunc = "mean")
```

```python
sales.pivot_table(index = "Date", aggfunc= "sum")
```

```python
sales.pivot_table(
    index = "Date", values = "Revenue", aggfunc = "sum"
)
```

```python
# 여러 열에 걸쳐 값을 집계하려면 리스트를 values에 전달하면 된다.
# 이제 날짜별로 그룹화된 수익 합계를 구했다. 마지막 단계는 각 영업 사원이 일일 총계에 기여한 금액을 구한다.
# 영업 사원의 이름을 별도의 열에 배치하면 데이터셋을 읽기 편하다.

# 행 축은 날짜별로 구성되고 열 축은 영업 사원의 이름으로 구성된 수이그이 합계 집계 완료

sales.pivot_table(
        index = "Date",
        columns = "Name",
        values = "Revenue",
        aggfunc = "sum"

)
```

```python
sales.pivot_table(
                index = "Date",
                columns = "Name",
                values = "Revenue",
                aggfunc = "sum",
                fill_value = 0

)
```

```python
# margins의 경우는 true를 전달하여 각 행과 열에 대한 합계를 구할 수 있다.
sales.pivot_table(
        index = "Date",
        columns = "Name",
        values = "Revenue",
        aggfunc = "sum",
        fill_value = 0,
        margins = True
)
```

```python
# 엥 근데 레이블에 'All'이 추가되면서 날짜의 표현 방식이 바뀌었다.
# 테이블을 'All'에서 'Total'로 변경하는 예제

sales.pivot_table(
            index = "Date",
            columns = "Name",
            values = "Revenue",
            aggfunc = "sum",
            fill_value = 0,
            margins = True,
            margins_name = "Total"

)
```

```python
sales.pivot_table(
        index = "Date",
        columns = "Name",
        values = "Revenue",
        aggfunc = "count"
)
```

```python
sales.pivot_table(
        index = "Date",
        columns = "Name",
        values = "Revenue",
        aggfunc = ["sum", "count"],
        fill_value = 0

)
```

```python
sales.pivot_table(
        index = "Date",
        columns = "Name",
        values = ["Revenue", "Expenses"],
        fill_value = 0,
        aggfunc = {"Revenue": "min", "Expenses": "min"}

)
```

```python
sales.pivot_table(
            index = ["Name", "Date"], values = "Revenue", aggfunc = "sum"

)
```

```python
cars = pd.read_csv("used_cars.csv")
cars
```

```python
min_wage = pd.read_csv("minimum_wage.csv")
min_wage.head()
```

```python
# 1. cars에서 가격의 합계를 집계해라. 행 축에서 연료 유형별로 결과를 그룹화해라
cars.pivot_table(index = "Fuel", values = "Price", aggfunc = "sum")
```

```python
# 2. Cars에서 자동차의 개수를 집계해라. 인덱스 축에서 제조업체별로 결과를 그룹화하고 열 축에서 변속기 유형을 그룹화해라. 
# 행과 열 모두에 대한 소계를 표시하자.
cars.pivot_table(index = "Manufacturer", columns = "Transmission", values = "Price", aggfunc = "count", margins = True, margins_name = "Total")
```

```python
# 3. Cars에서 자동차의 평균 가격을 집계해라. 인덱스 축에 연도 및 연료 유형별로 결과를 그룹화하고 열축에 변속기 유형을 그룹화한다.
cars.pivot_table(values = "Price", index = ["Year","Fuel"], columns = "Transmission"
                , aggfunc = "mean"
                )
```

```python
# 4. 3번 문제의 데이터프레임이 주어졌을 때 열 축에서 행 축으로 변속기 유형을 옮겨라
temp_cars = cars.pivot_table(values = "Price", index = ["Year","Fuel"], columns = "Transmission"
                , aggfunc = "mean")

temp_cars.stack()
```
