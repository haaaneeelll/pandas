```python
# 차원의 수는 자료구조에서 값을 추출할 때 필요한 참조점의 개수이다.
# 시리즈에서 값을 찾으려면 하나의 레이블이나 하나의 인덱스 위치가 필요하다.
# 데이터프레임에서 값을 찾으려면 행에 대한 레이블/인덱스와 열에 대한 레이블/인덱스라는 2개의 참조점이 필요하다.
# 판다스는 multiIndex를 사용하여 다차원의 데이터셋을 지원한다.
```

```python
import pandas as pd
```

```python
address = ("8809 Flari Square", "Toddsid", "IL", "37206")
address
```

```python
addresses = [
    ("8809 Flair Square", "Toddside", "IL", "37206"),
    ("9901 Austin Street", "Toddside", "IL", "37206"),
    ("905 Hogan Quarter", "Franklin", "IL", "37206"),
]
```

```python
addresses
```

```python
# 인덱스 레이블로 행을 참조할 수 있으며 각 인덱스 레이블이 여러 요소를 포함하는 컨테이너라는 차이만 있습니다.
# 이와 같이 멀티인덱스 객체는 각 레이블이 여러 데이터 조각을 저장할 수 있는 인덱스라고 이해하면 된다.
# 멀티인덱스 클래스는 판스 라이브러리에서 최상위 속성으로 접근할 수 있다.
# 멀티인덱스 클래스는 from_tuples 클래스 메서드를 제공하며 이 메서드를 활용하면 튜플리스트로부터 MultiIndex를 인스턴스화를 할 수 있다.
```

```python
pd.MultiIndex.from_tuples(addresses)
pd.MultiIndex.from_tuples(tuples = addresses)
```

```python
# 같은 위치에 있는 튜플 값의 모음을 판다스 용어로 MultiIndex의 레벨이라고 한다.
# 첫 번째 MultiIndex 레벨은 '8809 Flair Square', '9901 Austin Street', 'Hogan Quarter'
# 두 번째 MultiIndex 레벨은 'Toddside', 'Toddside', 'Franklin'
```

```python
row_index = pd.MultiIndex.from_tuples(
tuples = addresses,
names = ["Street", "City", "State", "Zip"]
)

row_index
```

```python
# 멀티인덱스는 각 레이블이 여러 값을 보유하는 인덱스 컨테이너이다.
# 레벨은 레이블에서 동일한 위치에 있는 값으로 구성된다.
# 멀티인덱스를 생성했으면 데이터프레임에 연결하겠다. 가장 쉬운 방법은 데이터프레임 생성자의 인덱스 매개변수를 사용하는 것이다.
# 멀티인덱스에는 3개의 튜플(또는 3개의 레이블)이 있으므로 3개의 데이터 행을 제공해야 한다.
```

# 정리
- 멀티인덱스는 여러 레벨로 구성된 인덱스다.
- 멀티인덱스는 값을 구성된 튜플을 사용하여 레이블을 저장한다.
- 데이터프레임은 행과 열 축 모두에 멀티인덱스를 저장할 수 있다.
- sort_index 머세드는 멀티인덱스 레벨을 정렬한다. 판다스는 인덱스 레벨을 개별적으로 또는 그룹으로 정렬할 수 있다.
- 레이블 기반의 Loc와 위치 기반의 iloc 접근자는 원하는 행과 열의 조합을 추출하려면 추가 인수가 필요하다.
- 튜플을 Loc와 iloc 접근자에게 전다랗면 구문의 모호성을 방지할 수 있다.
- reset_index 메서드는 인덱스 레벨을 데이터프레임 열로 통합한다.
- 기존 데이터프레임열에서 멀티인덱스를 만들고 싶다면 열 리스트를 set_index로 전달한다.
