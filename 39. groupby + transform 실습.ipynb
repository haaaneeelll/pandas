```python
import pandas as pd
pd.options.display.float_format = '{:.2f}'.format
df = pd.read_csv("https://raw.githubusercontent.com/"
                 "panda-kim/pandas/main/pokemon01.csv")
```

```python
df
```

# 1. 람다로 조건을 만든다.

```python
cond1 = df.groupby("타입")["체력"].transform(lambda x: x >= x.mean())
```

# 2. 람다조건으로 boolean masking

```python
df["분류"] = df["체력"].mask(cond1, "강인").mask(~cond1, "허약")
df
```
