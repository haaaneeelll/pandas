# 2. 여러 개의 웹페이지에서 1번을 반복해서 합친다.

```python
url = "https://sports.news.naver.com/kbaseball/record/index?category=kbo&year="
year = "2021"
 # years = ["2015","2016","2017", "2018", "2019", "2020", "2021"] 불편함
years = range(2015, 2022)
```

```python
df = pd.DataFrame([])
for i in years:
    df1 = pd.read_html(url + str(i))[0]
    df1["연도"] = str(i)
    df = pd.concat([df, df1])
df
    
```

# 3. 피벗 테이블 (1)

```python
df.pivot(values="순위", index="연도", columns="팀")
```

```python
df = df.replace({"kt":"KT", "SK":"SSG", "넥센":"키움"})
```
