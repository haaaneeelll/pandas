```python
import pandas as pd
df1= pd.read_excel("/Users/haneul/Desktop/실습/E19EXAMPLE.xlsx" ,sheet_name = 0)
df1
df = pd.DataFrame([[90,"A"],[80, "B"], [70, "C"], [0, "D"]], 
                  columns=["점수", "학점"]).sort_values("점수")
```

# 1. 등수매기기(내림차순, 동점자처리는 Min)

```python
df1["등수"] = df1["점수"].rank(ascending = False, method = "min")
df1
```

# 2. 학점부여(근사값 VLOOKUP방식으로)

```python
df2 = df1.sort_values(["점수"])
df2 = pd.merge_asof(df2, df)


```

# 3. 초기상태로 정렬 및 클립보드 복사

```python
df1.merge(df2, how = "left")
```
