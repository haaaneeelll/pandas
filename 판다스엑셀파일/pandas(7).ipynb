```python
import pandas as pd
```

```python
df1 = pd.read_excel('Data05.xlsx', skiprows = 1)
```

```python
df1
```

```python
df4 = pd.DataFrame(df1.iloc[:, 6:-2].drop(['Unnamed: 7', '판매'], axis=1).set_index('제품명').stack()).reset_index()
```

```python
df5 = df4.rename(columns = {'level_1': '날짜', 0:'재고량'})
```

```python
df1.head(2)
```

```python
df6 = df1[['카테고리명','자재그룹','자재그룹명','제품코드','안전재고','제품명',' 분류']] 
```

```python
df5
```

```python
df6
```

```python
df7 = df6.merge(df5, on='제품명', how='right')
df7.to_excel('total_Data.xlsx')
```

```python
df7
```

```python
df7['날짜(전처리)'] = pd.to_datetime(df7['날짜'])
```

```python
df7['공급월'] = df7['날짜(전처리)'].dt.month
df7['공급주'] = df7['날짜(전처리)'].dt.week
df7['공급일'] = df7['날짜(전처리)'].dt.day
```

```python
pd.pivot_table(data= df7, index = '공급월', values = '재고량', aggfunc = 'mean')
```
