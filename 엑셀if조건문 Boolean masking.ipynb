# Boolean masking

1. 조건문을 만든다.
2. 조건문을 적용할 자료를 부르고 Mask함수를 적용한다.

df.mask(cond, 참일때의 값)

## 1. 엑셀의 If기능 Mask

```python
cond1 = df1["점수"].isnull()
df1["참석여부"] = df1["점수"].mask(cond1, "x").mask(~cond1, "o")
df1
```

 ## 2. 엑셀의 다중if: 다중 Mask

```python
cond2 = df1["점수"] > 70
cond3 = df1["점수"] > 90
df1["성적"] = df1["점수"].mask(~cond2, "C").mask(cond2 , "B").mask(cond3, "A")
df1
```
