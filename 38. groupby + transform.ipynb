# Agg대신 transform사용하기
## 그룹화 시켜서 그룹함수로 집계만 하는 것이 아니라 열을 만들 때도 있음.

```python
import pandas as pd
pd.options.display.float_format = '{:.2f}'.format
a1 = ['송중기', '국어', 89]
a2 = ['김나현', '영어', 42]
a3 = ['박효신', '국어', 78]
a4 = ['김범수', '영어', 61]
a5 = ['권보아', '국어', 91]
a6 = ['한동근', '영어', 38]
df1 = pd.DataFrame([a1, a2, a3, a4, a5, a6], columns=["이름", "과목", "점수"])
df2 = df1.copy()
df3 = df1.copy()
s1 = pd.DataFrame([1, 2, 3, 4, 5, 6], columns=["s1"])
```

# Groupby와 Transform으로 집계함수 열만들기

```python
df1["과목별 평균점수"] = df1.groupby("과목")["점수"].transform("mean")
df1
```

# 1번으로 불리언인덱싱 해보기

```python
df1.loc[df1["점수"] > df1["과목별 평균점수"]]
```

# 3. agg + merge

```python
gdf = df2.groupby("과목")["점수"].mean()
df2 = df2.merge(gdf.rename("과목평균"), on="과목", how="left")
df2
```

# 4. transform으로 lambda 사용해보기 

```python
df3["m+s"] = df3.groupby("과목")["점수"].transform(lambda x: x.mean() + x.std())
df3
```
