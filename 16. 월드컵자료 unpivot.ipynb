```python
import pandas as pd
df1= pd.read_excel("/Users/haneul/Desktop/실습/E14EXAMPLE.xlsx")
df1
```

# 데이터프레임 자르기(슬라이싱)

```python
df2 = df1.iloc[:, [3,4,5,6]]
df2
```

# Stack + reset_index + 컬럼 이름 바꾸기

```python
df3 = df2.stack().reset_index()
# 위에 처럼 안쓰고 melt함수를 쓸 수도 있습니다.
# df1.melt(id_vars = ["개최국"], value_vars = ["1위", "2위", "3위", "4위"], var_name = "순위", value_name = "국가")

df3.columns = ["횟수", "순위", "국가"]
df3
```

# 피벗테이블

```python
df4 = df3.pivot_table("횟수", index = "국가", columns = "순위", aggfunc = "count" )
df4
```

# sort_values해서 내림차순으로..

```python
df4.sort_values(by=["1위", "2위", "3위", "4위"], ascending=False).iloc[:8].round(0).fillna(0).astype(int)


```
