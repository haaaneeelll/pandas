# 시리즈에 사용자 정의함수 적용
## def로 정의해서 apply

```python
def age_class(x):
    if 20<= x < 30:
        return "20대"
    if 30 <= x < 40:
        return "30대"
    if 40 <= x < 50:
        return "40대"
age_class(47)
```

```python
df1["분류1"] = df1["나이"].apply(age_class)
df1
```

## lambda

```python
df1['분류'] = df1['나이'].apply(lambda age: f"{age // 10}0대")

df1['분류2'] = df1.apply(lambda row: f"{(row['나이']//10)*10}{row['성별'][0]}" if row['성별'] in ['남자', '여자'] else "", axis=1)
df1
```
