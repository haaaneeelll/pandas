```python
import pandas as pd
a1 = ['송중기', '남', 1, 79, 93]
a2 = ['김나현', '여', 1, 85, 85]
a3 = ['권보아', '여', 1, 91, 78]
a4 = ['김범수', '남', 1, 97, 98]
a5 = ['이효리', '여', 2, 93, 79]
a6 = ['박효신', '남', 2, 78, 90]
a7 = ['한예슬', '여', 2, 87, 74]
a8 = ['민경훈', '남', 2, 99, 89]
col1 = ["", "성별", "반", "국어", "영어"]
df1 = pd.DataFrame([a1, a2, a3, a4, a5, a6, a7, a8], columns=col1).set_index("")
```

# 1. 단일열을 group by해서 단일열에 적용한다.

```python
df1.groupby("성별", as_index = False)["영어"].agg("mean")
# 데이터프레임으로 보려면 as_index를 false로..
```

# 2. 다수열을 group by해서 다수열에 적용한다.
# 3. agg로 멀티 집계함수를 쓴다.

```python
df1.groupby(["성별","반"])[["국어", "영어"]].agg("mean")
```

# 4. 피벗테이블과 groupby의 차이를 안다.

```python
df1.groupby(["성별","반"])[["국어", "영어"]].agg("mean")
```

```python
df1.pivot_table(["국어","영어"], index = ["성별", "반"], aggfunc = "mean")
# 피벗테이블의 경우 위의 groupby와 같다. 이 점에서 피벗테이블의 경우는 인덱스에 몰빵하면 groupby와 비슷하다.
```
